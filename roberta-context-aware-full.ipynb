```python
# ============================================================
# MODEL: RoBERTa Context-Aware FULL
# Input: context + utterance
# Dataset: context-full
# Figures folder + ZIP created after training
# ============================================================

import os
import json
import random
import shutil
import warnings
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from IPython.display import display

import torch
import torch.nn as nn
from torch.utils.data import Dataset, DataLoader

from transformers import AutoTokenizer, AutoModel, get_linear_schedule_with_warmup
from torch.optim import AdamW

from sklearn.metrics import (
    f1_score,
    precision_score,
    recall_score,
    classification_report
)

from tqdm.auto import tqdm

warnings.filterwarnings("ignore")


# ============================================================
# 1. Configuration
# ============================================================

class CFG:
    DATA_ROOT = Path("/kaggle/input/datasets/muhammadsaad003/cse495b/NLP dataset")

    # Context-aware FULL dataset
    TRAIN_FILE = DATA_ROOT / "context-full/train.jsonl"
    DEV_FILE   = DATA_ROOT / "context-full/dev.jsonl"
    TEST_FILE  = DATA_ROOT / "context-full/test.jsonl"
    LABEL_FILE = DATA_ROOT / "labels.txt"

    MODEL_NAME = "roberta-base"

    # Context-aware input needs longer length
    MAX_LEN = 512

    # For full dataset, batch size 4 is safer on Kaggle GPU
    BATCH_SIZE = 4

    # Gradient accumulation makes effective batch size = 4 * 2 = 8
    GRAD_ACCUM_STEPS = 2

    EPOCHS = 3

    LR = 2e-5
    WEIGHT_DECAY = 0.01
    WARMUP_RATIO = 0.10

    LABEL_THRESHOLD = 0.30
    DEFAULT_PRED_THRESHOLD = 0.30

    SEED = 42

    OUTPUT_DIR = Path("/kaggle/working/roberta_context_aware_full")

    FIGURE_DIR = OUTPUT_DIR / "RoBERTa_context_aware_full_figures"
    FIGURE_ZIP_NAME = OUTPUT_DIR / "RoBERTa_context_aware_full_figures"


CFG.OUTPUT_DIR.mkdir(parents=True, exist_ok=True)
CFG.FIGURE_DIR.mkdir(parents=True, exist_ok=True)

print("Train file:", CFG.TRAIN_FILE)
print("Dev file:", CFG.DEV_FILE)
print("Test file:", CFG.TEST_FILE)
print("Label file:", CFG.LABEL_FILE)
print("Output dir:", CFG.OUTPUT_DIR)
print("Figure dir:", CFG.FIGURE_DIR)


# ============================================================
# 2. Check Files
# ============================================================

required_files = [
    CFG.TRAIN_FILE,
    CFG.DEV_FILE,
    CFG.TEST_FILE,
    CFG.LABEL_FILE
]

for file_path in required_files:
    if not file_path.exists():
        raise FileNotFoundError(f"File not found: {file_path}")

print("\nAll required files found.")


# ============================================================
# 3. Reproducibility
# ============================================================

def set_seed(seed):
    random.seed(seed)
    np.random.seed(seed)

    torch.manual_seed(seed)
    torch.cuda.manual_seed_all(seed)

    torch.backends.cudnn.deterministic = False
    torch.backends.cudnn.benchmark = True


set_seed(CFG.SEED)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("\nDevice:", device)

if torch.cuda.is_available():
    print("GPU:", torch.cuda.get_device_name(0))


# ============================================================
# 4. Load Labels
# ============================================================

def load_labels(path):
    labels = []

    with open(path, "r", encoding="utf-8") as f:
        for line in f:
            label = line.strip()
            if label:
                labels.append(label)

    return labels


LABELS = load_labels(CFG.LABEL_FILE)
NUM_LABELS = len(LABELS)

id2label = {i: label for i, label in enumerate(LABELS)}
label2id = {label: i for i, label in enumerate(LABELS)}

print("\nNumber of labels:", NUM_LABELS)
print(LABELS)


# ============================================================
# 5. Inspect One Raw Record
# ============================================================

with open(CFG.TRAIN_FILE, "r", encoding="utf-8") as f:
    raw_sample = json.loads(f.readline())

print("\nRaw sample keys:")
print(raw_sample.keys())

print("\nRaw sample preview:")
print(json.dumps(raw_sample, indent=2, ensure_ascii=False)[:3000])


# ============================================================
# 6. Helper Functions
# ============================================================

def get_context(record):
    possible_keys = [
        "context",
        "summary",
        "clean_context",
        "cleaned_context",
        "cleaned_summary",
        "context_clean"
    ]

    for key in possible_keys:
        value = record.get(key, "")
        if isinstance(value, str) and value.strip():
            return value.strip()

    return ""


def get_utterance(record):
    """
    For this context-aware full experiment, we use original utterance.

    If you want rewritten utterance later, change this order:
    ["rewritten_utterance", "utterance", ...]
    """
    possible_keys = [
        "utterance",
        "original_utterance",
        "text",
        "sentence"
    ]

    for key in possible_keys:
        value = record.get(key, "")
        if isinstance(value, str) and value.strip():
            return value.strip()

    return ""


def get_label_indices(record):
    possible_keys = [
        "labels",
        "label",
        "emotions",
        "emotion_labels"
    ]

    for key in possible_keys:
        value = record.get(key, None)
        if value is not None:
            return value

    return []


def get_expressiveness(record):
    possible_keys = [
        "expressiveness",
        "scores",
        "label_scores",
        "emotion_scores"
    ]

    for key in possible_keys:
        value = record.get(key, None)
        if value is not None:
            return value

    return []


def make_multihot(label_values, expressiveness, num_labels, threshold=0.30):
    y = np.zeros(num_labels, dtype=np.float32)

    if label_values is None:
        return y

    if not isinstance(label_values, list):
        label_values = [label_values]

    if expressiveness is None or expressiveness == []:
        expressiveness = [1.0] * len(label_values)

    if not isinstance(expressiveness, list):
        expressiveness = [expressiveness] * len(label_values)

    if len(expressiveness) < len(label_values):
        expressiveness = expressiveness + [1.0] * (len(label_values) - len(expressiveness))

    for label_value, score in zip(label_values, expressiveness):
        try:
            score = float(score)
        except Exception:
            score = 1.0

        if score < threshold:
            continue

        idx = None

        if isinstance(label_value, int):
            idx = label_value

        elif isinstance(label_value, float):
            idx = int(label_value)

        elif isinstance(label_value, str):
            label_value_clean = label_value.strip()

            if label_value_clean.isdigit():
                idx = int(label_value_clean)
            elif label_value_clean in label2id:
                idx = label2id[label_value_clean]

        elif isinstance(label_value, dict):
            possible_label = (
                label_value.get("label")
                or label_value.get("emotion")
                or label_value.get("name")
                or label_value.get("id")
            )

            possible_score = (
                label_value.get("score")
                or label_value.get("expressiveness")
                or label_value.get("value")
            )

            if possible_score is not None:
                try:
                    score = float(possible_score)
                except Exception:
                    score = 1.0

            if score < threshold:
                continue

            if isinstance(possible_label, int):
                idx = possible_label
            elif isinstance(possible_label, str):
                if possible_label.isdigit():
                    idx = int(possible_label)
                elif possible_label in label2id:
                    idx = label2id[possible_label]

        if idx is not None and 0 <= idx < num_labels:
            y[idx] = 1.0

    return y


sample_y = make_multihot(
    get_label_indices(raw_sample),
    get_expressiveness(raw_sample),
    NUM_LABELS,
    threshold=CFG.LABEL_THRESHOLD
)

print("\nPositive labels in raw sample:")
print([LABELS[i] for i, v in enumerate(sample_y) if v == 1])


# ============================================================
# 7. Load Context-aware FULL JSONL Dataset
# ============================================================

def load_context_aware_jsonl(path, num_labels, threshold=0.30):
    rows = []

    with open(path, "r", encoding="utf-8") as f:
        for line in tqdm(f, desc=f"Loading {path.name}"):
            line = line.strip()

            if not line:
                continue

            record = json.loads(line)

            context = get_context(record)
            utterance = get_utterance(record)

            label_values = get_label_indices(record)
            expressiveness = get_expressiveness(record)

            y = make_multihot(
                label_values=label_values,
                expressiveness=expressiveness,
                num_labels=num_labels,
                threshold=threshold
            )

            if len(utterance.strip()) == 0:
                continue

            if y.sum() == 0:
                continue

            rows.append({
                "context": context,
                "utterance": utterance,
                "labels": y,
                "primary_emotion": record.get("primary_emotion", record.get("main_emotion", None)),
                "raw": record
            })

    df = pd.DataFrame(rows)
    return df


train_df = load_context_aware_jsonl(CFG.TRAIN_FILE, NUM_LABELS, CFG.LABEL_THRESHOLD)
dev_df   = load_context_aware_jsonl(CFG.DEV_FILE, NUM_LABELS, CFG.LABEL_THRESHOLD)
test_df  = load_context_aware_jsonl(CFG.TEST_FILE, NUM_LABELS, CFG.LABEL_THRESHOLD)

print("\nTrain shape:", train_df.shape)
print("Dev shape:", dev_df.shape)
print("Test shape:", test_df.shape)

print("\nProcessed sample:")
print("Context:", train_df.iloc[0]["context"][:500])
print("Utterance:", train_df.iloc[0]["utterance"])
print("Labels:", [LABELS[i] for i, v in enumerate(train_df.iloc[0]["labels"]) if v == 1])


# ============================================================
# 8. Load Tokenizer
# ============================================================

tokenizer = AutoTokenizer.from_pretrained(CFG.MODEL_NAME)


# ============================================================
# 9. Figure: Label Distribution
# ============================================================

def label_distribution(df, labels):
    label_matrix = np.stack(df["labels"].values)
    counts = label_matrix.sum(axis=0)

    dist = pd.DataFrame({
        "label": labels,
        "count": counts.astype(int)
    }).sort_values("count", ascending=False)

    return dist


dist_df = label_distribution(train_df, LABELS)
dist_df.to_csv(CFG.OUTPUT_DIR / "train_label_distribution.csv", index=False)

print("\nTraining label distribution:")
display(dist_df)

plt.figure(figsize=(12, 6))
plt.bar(dist_df["label"], dist_df["count"])
plt.xticks(rotation=90)
plt.xlabel("Emotion label")
plt.ylabel("Count")
plt.title("Training Label Distribution - RoBERTa Context-Aware Full")
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_label_distribution.png", dpi=300)
plt.show()


# ============================================================
# 10. Figures: Token Length Analysis
# ============================================================

def token_length(text):
    return len(tokenizer.encode(str(text), add_special_tokens=False))


train_df["context_token_len"] = train_df["context"].apply(token_length)
train_df["utterance_token_len"] = train_df["utterance"].apply(token_length)


def combined_token_length(row):
    encoded = tokenizer(
        str(row["context"]),
        str(row["utterance"]),
        add_special_tokens=True,
        truncation=False
    )
    return len(encoded["input_ids"])


train_df["combined_token_len"] = train_df.apply(combined_token_length, axis=1)

length_summary = train_df[
    ["context_token_len", "utterance_token_len", "combined_token_len"]
].describe()

length_summary.to_csv(CFG.OUTPUT_DIR / "token_length_summary.csv")

print("\nToken length summary:")
display(length_summary)

plt.figure(figsize=(8, 5))
plt.hist(train_df["context_token_len"], bins=50)
plt.xlabel("Context token length")
plt.ylabel("Frequency")
plt.title("Context Token Length Distribution")
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_context_token_length.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(train_df["utterance_token_len"], bins=50)
plt.xlabel("Utterance token length")
plt.ylabel("Frequency")
plt.title("Utterance Token Length Distribution")
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_utterance_token_length.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.hist(train_df["combined_token_len"], bins=50)
plt.axvline(CFG.MAX_LEN, linestyle="--")
plt.xlabel("Context + utterance token length")
plt.ylabel("Frequency")
plt.title("Combined Token Length Distribution")
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_combined_token_length.png", dpi=300)
plt.show()

num_truncated = int((train_df["combined_token_len"] > CFG.MAX_LEN).sum())
print(f"\nExamples longer than MAX_LEN={CFG.MAX_LEN}: {num_truncated}")


# ============================================================
# 11. Dataset and DataLoader
# ============================================================

class ContextAwareEmotionDataset(Dataset):
    def __init__(self, df, tokenizer, max_len):
        self.contexts = df["context"].tolist()
        self.utterances = df["utterance"].tolist()
        self.labels = np.stack(df["labels"].values)
        self.tokenizer = tokenizer
        self.max_len = max_len

    def __len__(self):
        return len(self.utterances)

    def __getitem__(self, idx):
        context = str(self.contexts[idx])
        utterance = str(self.utterances[idx])
        labels = self.labels[idx]

        encoded = self.tokenizer(
            context,
            utterance,
            add_special_tokens=True,
            max_length=self.max_len,
            padding="max_length",
            truncation=True,
            return_attention_mask=True,
            return_tensors="pt"
        )

        return {
            "input_ids": encoded["input_ids"].squeeze(0),
            "attention_mask": encoded["attention_mask"].squeeze(0),
            "labels": torch.tensor(labels, dtype=torch.float)
        }


train_dataset = ContextAwareEmotionDataset(train_df, tokenizer, CFG.MAX_LEN)
dev_dataset   = ContextAwareEmotionDataset(dev_df, tokenizer, CFG.MAX_LEN)
test_dataset  = ContextAwareEmotionDataset(test_df, tokenizer, CFG.MAX_LEN)

train_loader = DataLoader(
    train_dataset,
    batch_size=CFG.BATCH_SIZE,
    shuffle=True,
    num_workers=2,
    pin_memory=True
)

dev_loader = DataLoader(
    dev_dataset,
    batch_size=CFG.BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=CFG.BATCH_SIZE,
    shuffle=False,
    num_workers=2,
    pin_memory=True
)


# ============================================================
# 12. Model
# ============================================================

class RobertaContextAwareEmotionClassifier(nn.Module):
    def __init__(self, model_name, num_labels, dropout=0.2):
        super().__init__()

        self.encoder = AutoModel.from_pretrained(model_name)
        hidden_size = self.encoder.config.hidden_size

        self.dropout = nn.Dropout(dropout)
        self.classifier = nn.Linear(hidden_size, num_labels)

    def forward(self, input_ids, attention_mask):
        outputs = self.encoder(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        cls_representation = outputs.last_hidden_state[:, 0, :]
        cls_representation = self.dropout(cls_representation)

        logits = self.classifier(cls_representation)

        return logits


model = RobertaContextAwareEmotionClassifier(
    model_name=CFG.MODEL_NAME,
    num_labels=NUM_LABELS,
    dropout=0.2
)

model = model.to(device)


# ============================================================
# 13. Loss, Optimizer, Scheduler
# ============================================================

criterion = nn.BCEWithLogitsLoss()

optimizer = AdamW(
    model.parameters(),
    lr=CFG.LR,
    weight_decay=CFG.WEIGHT_DECAY
)

total_update_steps = (len(train_loader) // CFG.GRAD_ACCUM_STEPS) * CFG.EPOCHS
warmup_steps = int(total_update_steps * CFG.WARMUP_RATIO)

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=warmup_steps,
    num_training_steps=total_update_steps
)

scaler = torch.cuda.amp.GradScaler(enabled=torch.cuda.is_available())

print("\nNumber of batches per epoch:", len(train_loader))
print("Gradient accumulation steps:", CFG.GRAD_ACCUM_STEPS)
print("Total optimizer update steps:", total_update_steps)
print("Warmup steps:", warmup_steps)


# ============================================================
# 14. Metrics
# ============================================================

def compute_metrics(y_true, y_prob, threshold=0.30):
    y_pred = (y_prob >= threshold).astype(int)

    macro_f1 = f1_score(y_true, y_pred, average="macro", zero_division=0)
    micro_f1 = f1_score(y_true, y_pred, average="micro", zero_division=0)

    macro_precision = precision_score(y_true, y_pred, average="macro", zero_division=0)
    macro_recall = recall_score(y_true, y_pred, average="macro", zero_division=0)

    return {
        "macro_f1": macro_f1,
        "micro_f1": micro_f1,
        "macro_precision": macro_precision,
        "macro_recall": macro_recall
    }


def threshold_search_df(y_true, y_prob):
    rows = []

    for threshold in np.arange(0.05, 0.96, 0.01):
        y_pred = (y_prob >= threshold).astype(int)

        rows.append({
            "threshold": float(threshold),
            "macro_f1": f1_score(y_true, y_pred, average="macro", zero_division=0),
            "macro_precision": precision_score(y_true, y_pred, average="macro", zero_division=0),
            "macro_recall": recall_score(y_true, y_pred, average="macro", zero_division=0)
        })

    return pd.DataFrame(rows)


def find_best_threshold(y_true, y_prob):
    th_df = threshold_search_df(y_true, y_prob)
    best_row = th_df.sort_values("macro_f1", ascending=False).iloc[0]

    return float(best_row["threshold"]), float(best_row["macro_f1"]), th_df


# ============================================================
# 15. Training and Evaluation Functions
# ============================================================

def train_one_epoch(model, loader, optimizer, scheduler, criterion, device):
    model.train()

    total_loss = 0.0
    optimizer.zero_grad()

    progress = tqdm(loader, desc="Training", leave=False)

    for step, batch in enumerate(progress):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        with torch.cuda.amp.autocast(enabled=torch.cuda.is_available()):
            logits = model(
                input_ids=input_ids,
                attention_mask=attention_mask
            )

            loss = criterion(logits, labels)

            # Normalize loss for gradient accumulation
            loss = loss / CFG.GRAD_ACCUM_STEPS

        scaler.scale(loss).backward()

        if (step + 1) % CFG.GRAD_ACCUM_STEPS == 0 or (step + 1) == len(loader):
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)

            scaler.step(optimizer)
            scaler.update()

            optimizer.zero_grad()
            scheduler.step()

        total_loss += loss.item() * CFG.GRAD_ACCUM_STEPS

        progress.set_postfix({
            "loss": loss.item() * CFG.GRAD_ACCUM_STEPS
        })

    avg_loss = total_loss / len(loader)

    return avg_loss


@torch.no_grad()
def evaluate(model, loader, criterion, device, threshold=0.30):
    model.eval()

    total_loss = 0.0

    all_probs = []
    all_labels = []

    progress = tqdm(loader, desc="Evaluating", leave=False)

    for batch in progress:
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)

        logits = model(
            input_ids=input_ids,
            attention_mask=attention_mask
        )

        loss = criterion(logits, labels)

        probs = torch.sigmoid(logits)

        total_loss += loss.item()

        all_probs.append(probs.cpu().numpy())
        all_labels.append(labels.cpu().numpy())

    y_prob = np.vstack(all_probs)
    y_true = np.vstack(all_labels)

    metrics = compute_metrics(
        y_true=y_true,
        y_prob=y_prob,
        threshold=threshold
    )

    avg_loss = total_loss / len(loader)

    return avg_loss, metrics, y_true, y_prob


# ============================================================
# 16. Training Loop
# ============================================================

best_macro_f1 = -1.0
best_model_path = CFG.OUTPUT_DIR / "best_model.pt"

history = []

for epoch in range(1, CFG.EPOCHS + 1):
    print(f"\n================ Epoch {epoch}/{CFG.EPOCHS} ================")

    train_loss = train_one_epoch(
        model=model,
        loader=train_loader,
        optimizer=optimizer,
        scheduler=scheduler,
        criterion=criterion,
        device=device
    )

    dev_loss, dev_metrics, dev_y_true, dev_y_prob = evaluate(
        model=model,
        loader=dev_loader,
        criterion=criterion,
        device=device,
        threshold=CFG.DEFAULT_PRED_THRESHOLD
    )

    print(f"Train loss: {train_loss:.4f}")
    print(f"Dev loss:   {dev_loss:.4f}")

    for metric_name, metric_value in dev_metrics.items():
        print(f"{metric_name}: {metric_value:.4f}")

    row = {
        "epoch": epoch,
        "train_loss": train_loss,
        "dev_loss": dev_loss,
        **dev_metrics
    }

    history.append(row)

    if dev_metrics["macro_f1"] > best_macro_f1:
        best_macro_f1 = dev_metrics["macro_f1"]

        torch.save(model.state_dict(), best_model_path)

        print(f"Saved best model to: {best_model_path}")
        print(f"Best dev macro F1: {best_macro_f1:.4f}")


history_df = pd.DataFrame(history)
history_df.to_csv(CFG.OUTPUT_DIR / "training_history.csv", index=False)

print("\nTraining finished.")
display(history_df)


# ============================================================
# 17. Figures: Training Curves
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(history_df["epoch"], history_df["train_loss"], marker="o", label="train_loss")
plt.plot(history_df["epoch"], history_df["dev_loss"], marker="o", label="dev_loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training and Dev Loss - RoBERTa Context-Aware Full")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_loss_curve.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_df["epoch"], history_df["macro_f1"], marker="o", label="macro_f1")
plt.plot(history_df["epoch"], history_df["micro_f1"], marker="o", label="micro_f1")
plt.xlabel("Epoch")
plt.ylabel("F1-score")
plt.title("Dev F1 Curve - RoBERTa Context-Aware Full")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_dev_f1_curve.png", dpi=300)
plt.show()

plt.figure(figsize=(8, 5))
plt.plot(history_df["epoch"], history_df["macro_precision"], marker="o", label="macro_precision")
plt.plot(history_df["epoch"], history_df["macro_recall"], marker="o", label="macro_recall")
plt.xlabel("Epoch")
plt.ylabel("Score")
plt.title("Dev Precision and Recall Curve")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_dev_precision_recall_curve.png", dpi=300)
plt.show()


# ============================================================
# 18. Load Best Model and Tune Threshold
# ============================================================

model.load_state_dict(torch.load(best_model_path, map_location=device))
model = model.to(device)

dev_loss, dev_metrics_default, dev_y_true, dev_y_prob = evaluate(
    model=model,
    loader=dev_loader,
    criterion=criterion,
    device=device,
    threshold=CFG.DEFAULT_PRED_THRESHOLD
)

best_threshold, best_dev_f1, threshold_df = find_best_threshold(
    y_true=dev_y_true,
    y_prob=dev_y_prob
)

threshold_df.to_csv(CFG.OUTPUT_DIR / "threshold_search.csv", index=False)

with open(CFG.OUTPUT_DIR / "best_threshold.txt", "w") as f:
    f.write(str(best_threshold))

print("\nBest threshold:", best_threshold)
print("Best dev macro F1:", best_dev_f1)

display(threshold_df.sort_values("macro_f1", ascending=False).head(10))


# ============================================================
# 19. Figure: Threshold Search
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(threshold_df["threshold"], threshold_df["macro_f1"], label="macro_f1")
plt.plot(threshold_df["threshold"], threshold_df["macro_precision"], label="macro_precision")
plt.plot(threshold_df["threshold"], threshold_df["macro_recall"], label="macro_recall")
plt.axvline(best_threshold, linestyle="--", label=f"best={best_threshold:.2f}")
plt.xlabel("Prediction threshold")
plt.ylabel("Score")
plt.title("Threshold Search on Dev Set")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_threshold_search.png", dpi=300)
plt.show()


# ============================================================
# 20. Final Test Evaluation
# ============================================================

test_loss, test_metrics_default, test_y_true, test_y_prob = evaluate(
    model=model,
    loader=test_loader,
    criterion=criterion,
    device=device,
    threshold=CFG.DEFAULT_PRED_THRESHOLD
)

test_metrics_best = compute_metrics(
    y_true=test_y_true,
    y_prob=test_y_prob,
    threshold=best_threshold
)

print("\n================ FINAL TEST RESULTS ================")
print(f"Test loss: {test_loss:.4f}")

print("\nUsing default threshold:", CFG.DEFAULT_PRED_THRESHOLD)
for metric_name, metric_value in test_metrics_default.items():
    print(f"{metric_name}: {metric_value:.4f}")

print("\nUsing best dev threshold:", round(best_threshold, 3))
for metric_name, metric_value in test_metrics_best.items():
    print(f"{metric_name}: {metric_value:.4f}")


result_df = pd.DataFrame([
    {
        "model": "roberta_context_aware_full",
        "input": "context + utterance",
        "dataset": "context-full",
        "threshold_type": "default",
        "threshold": CFG.DEFAULT_PRED_THRESHOLD,
        **test_metrics_default
    },
    {
        "model": "roberta_context_aware_full",
        "input": "context + utterance",
        "dataset": "context-full",
        "threshold_type": "best_dev",
        "threshold": best_threshold,
        **test_metrics_best
    }
])

result_df.to_csv(CFG.OUTPUT_DIR / "test_results.csv", index=False)

print("\nTest result table:")
display(result_df)


# ============================================================
# 21. Classification Report
# ============================================================

test_y_pred = (test_y_prob >= best_threshold).astype(int)

report = classification_report(
    test_y_true,
    test_y_pred,
    target_names=LABELS,
    zero_division=0,
    output_dict=True
)

report_df = pd.DataFrame(report).transpose()
report_df.to_csv(CFG.OUTPUT_DIR / "classification_report.csv")

print("\nClassification report:")
display(report_df)


# ============================================================
# 22. Figure: Per-Class F1
# ============================================================

per_class_df = report_df.loc[LABELS].reset_index()
per_class_df = per_class_df.rename(columns={"index": "label"})
per_class_df = per_class_df.sort_values("f1-score", ascending=False)

per_class_df.to_csv(CFG.OUTPUT_DIR / "per_class_scores.csv", index=False)

plt.figure(figsize=(12, 6))
plt.bar(per_class_df["label"], per_class_df["f1-score"])
plt.xticks(rotation=90)
plt.xlabel("Emotion label")
plt.ylabel("F1-score")
plt.title("Per-Class F1 - RoBERTa Context-Aware Full")
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_per_class_f1.png", dpi=300)
plt.show()


# ============================================================
# 23. Figure: Per-Class Precision and Recall
# ============================================================

plt.figure(figsize=(12, 6))
x = np.arange(len(per_class_df))
width = 0.4

plt.bar(x - width / 2, per_class_df["precision"], width, label="precision")
plt.bar(x + width / 2, per_class_df["recall"], width, label="recall")

plt.xticks(x, per_class_df["label"], rotation=90)
plt.xlabel("Emotion label")
plt.ylabel("Score")
plt.title("Per-Class Precision and Recall - RoBERTa Context-Aware Full")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_per_class_precision_recall.png", dpi=300)
plt.show()


# ============================================================
# 24. Figure: True vs Predicted Label Counts
# ============================================================

true_counts = test_y_true.sum(axis=0)
pred_counts = test_y_pred.sum(axis=0)

count_compare_df = pd.DataFrame({
    "label": LABELS,
    "true_count": true_counts.astype(int),
    "predicted_count": pred_counts.astype(int)
}).sort_values("true_count", ascending=False)

count_compare_df.to_csv(CFG.OUTPUT_DIR / "true_vs_predicted_counts.csv", index=False)

print("\nTrue vs predicted label counts:")
display(count_compare_df)

x = np.arange(len(count_compare_df))
width = 0.4

plt.figure(figsize=(12, 6))
plt.bar(x - width / 2, count_compare_df["true_count"], width, label="true")
plt.bar(x + width / 2, count_compare_df["predicted_count"], width, label="predicted")
plt.xticks(x, count_compare_df["label"], rotation=90)
plt.xlabel("Emotion label")
plt.ylabel("Count")
plt.title("True vs Predicted Label Counts - RoBERTa Context-Aware Full")
plt.legend()
plt.tight_layout()
plt.savefig(CFG.FIGURE_DIR / "figure_true_vs_predicted_counts.png", dpi=300)
plt.show()


# ============================================================
# 25. Save Model Config and Tokenizer
# ============================================================

tokenizer.save_pretrained(CFG.OUTPUT_DIR)

config = {
    "model_name": CFG.MODEL_NAME,
    "model_type": "roberta_context_aware_full",
    "input_format": "context + utterance",
    "dataset": "context-full",
    "num_labels": NUM_LABELS,
    "labels": LABELS,
    "label2id": label2id,
    "id2label": id2label,
    "max_len": CFG.MAX_LEN,
    "best_threshold": best_threshold,
    "label_threshold": CFG.LABEL_THRESHOLD,
    "batch_size": CFG.BATCH_SIZE,
    "grad_accum_steps": CFG.GRAD_ACCUM_STEPS,
    "effective_batch_size": CFG.BATCH_SIZE * CFG.GRAD_ACCUM_STEPS,
    "epochs": CFG.EPOCHS,
    "learning_rate": CFG.LR
}

with open(CFG.OUTPUT_DIR / "config.json", "w", encoding="utf-8") as f:
    json.dump(config, f, indent=2, ensure_ascii=False)

print("\nModel/tokenizer/config saved.")


# ============================================================
# 26. Inference Function
# ============================================================

@torch.no_grad()
def predict_emotions_context_aware(
    context,
    utterance,
    model,
    tokenizer,
    labels,
    threshold=None,
    top_k=8
):
    model.eval()

    if threshold is None:
        threshold = best_threshold

    encoded = tokenizer(
        str(context),
        str(utterance),
        add_special_tokens=True,
        max_length=CFG.MAX_LEN,
        padding="max_length",
        truncation=True,
        return_attention_mask=True,
        return_tensors="pt"
    )

    input_ids = encoded["input_ids"].to(device)
    attention_mask = encoded["attention_mask"].to(device)

    logits = model(
        input_ids=input_ids,
        attention_mask=attention_mask
    )

    probs = torch.sigmoid(logits).cpu().numpy()[0]

    selected = []

    for i, p in enumerate(probs):
        if p >= threshold:
            selected.append({
                "emotion": labels[i],
                "score": float(p)
            })

    selected = sorted(selected, key=lambda x: x["score"], reverse=True)

    top_predictions = sorted(
        [
            {
                "emotion": labels[i],
                "score": float(probs[i])
            }
            for i in range(len(labels))
        ],
        key=lambda x: x["score"],
        reverse=True
    )[:top_k]

    return {
        "selected_emotions": selected,
        "top_predictions": top_predictions
    }


example_context = "A student has waited many months for an important result and finally opens the email."
example_utterance = "I cannot believe this finally happened."

prediction = predict_emotions_context_aware(
    context=example_context,
    utterance=example_utterance,
    model=model,
    tokenizer=tokenizer,
    labels=LABELS,
    threshold=best_threshold,
    top_k=8
)

print("\nExample context:")
print(example_context)

print("\nExample utterance:")
print(example_utterance)

print("\nSelected emotions:")
print(prediction["selected_emotions"])

print("\nTop predictions:")
print(prediction["top_predictions"])


# ============================================================
# 27. Save Test Predictions
# ============================================================

def create_prediction_df(df, y_prob, threshold, labels):
    rows = []

    for i in range(len(df)):
        probs = y_prob[i]
        pred_indices = np.where(probs >= threshold)[0]

        pred_labels = [labels[j] for j in pred_indices]
        pred_scores = [float(probs[j]) for j in pred_indices]

        true_indices = np.where(df.iloc[i]["labels"] == 1)[0]
        true_labels = [labels[j] for j in true_indices]

        rows.append({
            "context": df.iloc[i]["context"],
            "utterance": df.iloc[i]["utterance"],
            "true_primary_emotion": df.iloc[i]["primary_emotion"],
            "true_labels": true_labels,
            "predicted_labels": pred_labels,
            "predicted_scores": pred_scores
        })

    return pd.DataFrame(rows)


pred_df = create_prediction_df(
    df=test_df,
    y_prob=test_y_prob,
    threshold=best_threshold,
    labels=LABELS
)

pred_path = CFG.OUTPUT_DIR / "test_predictions.csv"
pred_df.to_csv(pred_path, index=False)

print("\nSaved predictions to:", pred_path)
display(pred_df.head(10))


# ============================================================
# 28. Zip All Generated Figures
# ============================================================

figure_zip_path = shutil.make_archive(
    base_name=str(CFG.FIGURE_ZIP_NAME),
    format="zip",
    root_dir=str(CFG.FIGURE_DIR)
)

print("\nAll figures saved in folder:")
print(CFG.FIGURE_DIR)

print("\nFigure ZIP file created:")
print(figure_zip_path)

print("\nFiles inside figure folder:")
for file_name in os.listdir(CFG.FIGURE_DIR):
    print("-", file_name)

print("\nFiles inside output folder:")
for file_name in os.listdir(CFG.OUTPUT_DIR):
    print("-", file_name)
```
